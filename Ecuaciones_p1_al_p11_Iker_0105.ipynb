#### problema 3
Paso: Sumamos los coeficientes de $x$ (términos semejantes).$7x = 21$Paso: Dividimos entre 7.$x = 21 / 7$Resultado: $x = 3$

3x+4x = 21

#### problema 5

6a - 2a + a = 20

```python
# Entradas manuales para simplificar los términos
c1 = float(input("Coeficiente 1 (ej. 6): "))
c2 = float(input("Coeficiente 2 (ej. -2): "))
c3 = float(input("Coeficiente 3 (ej. 1): "))
total = float(input("¿A qué es igual la suma? (ej. 20): "))
print(f"El valor de a es: {total / (c1 + c2 + c3)}")
```

#### problema 7
Paso: Agrupamos las $x$ a la izquierda restando $2x$, y los números a la derecha restando 3.$4x - 2x = 11 - 3$$2x = 8$Paso: Dividimos entre 2.$x = 8 / 2$Resultado: $x = 4$

4x +3 = 2x + 11

```python
a1 = float(input("Coeficiente de x a la izquierda (ej. 4): "))
b1 = float(input("Constante a la izquierda (ej. 3): "))
a2 = float(input("Coeficiente de x a la derecha (ej. 2): "))
b2 = float(input("Constante a la derecha (ej. 11): "))
# (a1 - a2)x = (b2 - b1)
x = (b2 - b1) / (a1 - a2)
print(f"El valor de x es: {x}")
```

#### problema 9
Paso: Agrupamos las $m$ a la izquierda restando $5m$, y los números a la derecha restando 4.$9m - 5m = 20 - 4$$4m = 16$Paso: Dividimos entre 4.$m = 16 / 4$Resultado: $m = 4$

9m + 4 = 5m + 20

```python
# Captura de datos para: 9m + 4 = 5m + 20
a1 = float(input("Coeficiente de m a la izquierda (9): "))
b1 = float(input("Constante a la izquierda (4): "))
a2 = float(input("Coeficiente de m a la derecha (5): "))
b2 = float(input("Constante a la derecha (20): "))

# Despeje: m = (b2 - b1) / (a1 - a2)
m = (b2 - b1) / (a1 - a2)
print(f"El valor de m es: {m}")
```

#### problema 11
Parece que hubo un pequeño error de dedo al escribir la ecuación: $6a + 5 = 2a + 21$. Asumiré que el "a" antes de "2a" fue un error y la ecuación correcta es la que hemos estado trabajando.Aquí tienes el procedimiento y el código solicitado.Procedimiento paso a paso ($6a + 5 = 2a + 21$)El objetivo es dejar todos los términos con $a$ de un lado y los números constantes del otro.Transposición de variables ($2a$): Restamos $2a$ en ambos lados para que las variables estén juntas en el lado izquierdo.$$6a - 2a + 5 = 21$$$$4a + 5 = 21$$Transposición de constantes ($5$): Restamos $5$ en ambos lados para pasar el número al lado derecho.$$4a = 21 - 5$$$$4a = 16$$Despeje de $a$: Como el $4$ multiplica a la $a$, dividimos ambos lados entre $4$.$$a = \frac{16}{4}$$$$a = 4$$

```python
print("Resolviendo: 6a + 5 = 2a + 21")

# Captura de los valores de la ecuación
a1 = float(input("Coeficiente de 'a' a la izquierda (6): "))
b1 = float(input("Constante a la izquierda (5): "))
a2 = float(input("Coeficiente de 'a' a la derecha (2): "))
b2 = float(input("Constante a la derecha (21): "))

# Procedimiento interno:
# (a1 - a2) * a = (b2 - b1)
# a = (b2 - b1) / (a1 - a2)

a = (b2 - b1) / (a1 - a2)

print(f"\nResultado paso 1 (agrupación): {a1-a2}a = {b2-b1}")
print(f"El valor final de 'a' es: {a}")
```
