# problema 1
$x + 8 = 15$: Es una ecuación de primer grado simple. Se resuelve restando 8 en ambos lados para dejar a la variable $x$ sola

```python
x_mas = float(input("¿Qué número se suma a x? (ej. 8): "))
igualdad = float(input("¿A qué es igual la suma? (ej. 15): "))
print(f"El valor de x es: {igualdad - x_mas}")
```

#### problema 3
$3x + 4x = 21$: Requiere simplificación de términos semejantes. Primero sumas las $x$ ($7x = 21$) y luego divides por el coeficiente

```python
coeff1 = float(input("Primer coeficiente de x (ej. 3): "))
coeff2 = float(input("Segundo coeficiente de x (ej. 4): "))
resultado = float(input("¿A qué es igual la suma? (ej. 21): "))
print(f"El valor de x es: {resultado / (coeff1 + coeff2)}")
```

#### problema 5
$6a - 2a + a = 20$: Igual que el anterior, agrupas los términos con la misma letra ($6-2+1 = 5a$) para llegar a $5a = 20$.

```python
# Entradas manuales para simplificar los términos
c1 = float(input("Coeficiente 1 (ej. 6): "))
c2 = float(input("Coeficiente 2 (ej. -2): "))
c3 = float(input("Coeficiente 3 (ej. 1): "))
total = float(input("¿A qué es igual la suma? (ej. 20): "))
print(f"El valor de a es: {total / (c1 + c2 + c3)}")
```

#### problema 7
$4x + 3 = 2x + 11$: Ecuación con incógnitas en ambos lados. Mueves las variables a la izquierda ($4x - 2x$) y los números a la derecha ($11 - 3$) para igualar

```python
a1 = float(input("Coeficiente de x a la izquierda (ej. 4): "))
b1 = float(input("Constante a la izquierda (ej. 3): "))
a2 = float(input("Coeficiente de x a la derecha (ej. 2): "))
b2 = float(input("Constante a la derecha (ej. 11): "))
# (a1 - a2)x = (b2 - b1)
x = (b2 - b1) / (a1 - a2)
print(f"El valor de x es: {x}")
```

#### problema 9
$9m + 4 = 5m + 20$: Misma lógica que el anterior; transposición de términos para agrupar variables y constantes

```python
# Captura de datos para: 9m + 4 = 5m + 20
a1 = float(input("Coeficiente de m a la izquierda (9): "))
b1 = float(input("Constante a la izquierda (4): "))
a2 = float(input("Coeficiente de m a la derecha (5): "))
b2 = float(input("Constante a la derecha (20): "))

# Despeje: m = (b2 - b1) / (a1 - a2)
m = (b2 - b1) / (a1 - a2)
print(f"El valor de m es: {m}")
```

#### problema 11
$6a + 5 = 2a + 21$: Nuevamente, reducción de términos mediante operaciones inversas hasta despejar la variable $a$.

```python
# Captura de datos para: 6a + 5 = 2a + 21
a1 = float(input("Coeficiente de a a la izquierda (6): "))
b1 = float(input("Constante a la izquierda (5): "))
a2 = float(input("Coeficiente de a a la derecha (2): "))
b2 = float(input("Constante a la derecha (21): "))

# Despeje: a = (b2 - b1) / (a1 - a2)
a = (b2 - b1) / (a1 - a2)
print(f"El valor de a es: {a}")
```
